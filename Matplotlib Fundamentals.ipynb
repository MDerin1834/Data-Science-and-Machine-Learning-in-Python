```python
import matplotlib.pyplot as plt  # Importing the library to create plots
import numpy as np  
# Importing numpy to work with arrays and mathematical functions
import pandas as pd  # Importing pandas to handle data from CSV files
import warnings # filter warnings
warnings.filterwarnings("ignore")

# EXAMPLE 1 - Simple line plot

# Define x, y, and z values
x = [1, 2, 3, 4]
y = [1, 8, 27, 64]  # y values represent cube of x
z = [3, 9, 34, 72]  # z values are arbitrary

plt.figure(figsize=(6, 3))  # Set figure size

# Plot the first line (x vs. y), purple color with line width of 3
plt.plot(x, y, color="purple", linewidth=3)

# Plot the second line (x vs. z), green dashed line with circle markers 
plt.plot(x, z, "o--g", linewidth=6)

# Set the axis limits: x-axis (0 to 5) and y-axis (0 to 75)
plt.axis([0, 5, 0, 75])

# Add title and labels to the plot
plt.title("What is this graph?")  # Title of the graph
plt.xlabel("numss")  # Label for the x-axis
plt.ylabel("numss2")  # Label for the y-axis

# Show the plot
plt.show()
```

```python
# EXAMPLE 2 - Plotting multiple functions 
# (linear, quadratic, cubic) on the same graph

# Generate 100 evenly spaced numbers between 0 and 6 using np.linspace
x = np.linspace(0, 6, 100)

plt.figure(figsize=(6, 3))  # Set figure size
# Set axis limits: x-axis (0 to 6), y-axis (0 to 220)
plt.axis([0, 6, 0, 220])

# Plot three functions (linear, quadratic, cubic) with labels
plt.plot(x, x, label="linear")      # Linear function y = x
plt.plot(x, x**2, label="quadratic")  # Quadratic function y = x^2
plt.plot(x, x**3, label="cubic")     # Cubic function y = x^3

# Display the legend to distinguish the different lines
plt.legend()
plt.show() # Show the plot
```

```python
# EXAMPLE 3 - Subplots (multiple plots in a grid)

# Generate 100 evenly spaced numbers between 0 and 6
x = np.linspace(0, 6, 100)

# Create a figure with a 2x2 grid of subplots (4 plots total)
fig, axs = plt.subplots(2, 2)

# First plot in position [0, 0] (top-left)
axs[0, 0].plot(x, x, color="red")  # Linear function
axs[0, 0].set_title("Plot 1")  # Set title for the first plot

# Second plot in position [0, 1] (top-right)
axs[0, 1].plot(x, x**2, color="yellow")  # Quadratic function
axs[0, 1].set_title("Plot 2")  # Set title for the second plot

# Third plot in position [1, 0] (bottom-left)
axs[1, 0].plot(x, x**3, color="green")  # Cubic function
axs[1, 0].set_title("Plot 3")  # Set title for the third plot

# Fourth plot in position [1, 1] (bottom-right)
axs[1, 1].plot(x, x**(1/2), color="purple")  # Square root function
axs[1, 1].plot(x, x**(1/3), color="pink")  # Cube root function
axs[1, 1].set_title("Plot 4")  # Set title for the fourth plot

# Add another line (inverse square root) to the second plot [0, 1]
axs[0, 1].plot(x, x**(-1/2), color="orange")

# Adjust the layout of the plots to prevent overlapping
plt.tight_layout()

# Show the subplots
plt.show()
```

```python
# EXAMPLE 4 - Plotting data from a CSV file using pandas

# Load data from the CSV file into a DataFrame
df = pd.read_csv("GBvideos.csv")

# Select only the relevant columns: title, likes, views, dislikes
# Group by 'title' and calculate the mean of 'likes', 'views', and 'dislikes'
df = df[["title", "likes", "views", 
         "dislikes"]].groupby("title").mean().head(10)

# Plot each column (likes, views, dislikes) in separate subplots
df.plot(subplots=True)

# Display the legend for each subplot
plt.legend()
plt.show() # Show the plots
```

```python
# Generate 20 evenly spaced numbers between -10 and 9
x = np.linspace(-10, 9, 20)

# Define two mathematical functions: cube and fourth power
y = x**3  # Cube function
z = x**4  # Fourth power function


# EXAMPLE 1 - Plot with inset (two plots in one figure)

# Create a new figure for the plot
figure = plt.figure(figsize=(6,3))

# Add a large set of axes to the figure (main plot)
# The list specifies [left, bottom, width, height] of the axes within the figure
axes = figure.add_axes([0.1, 0.1, 0.85, 0.85])
axes.plot(x, y, "b")  # Plot the cube function (y = x^3) in blue ("b")

# Label the axes and set a title for the main plot
axes.set_xlabel("X values")
axes.set_ylabel("Y values")
axes.set_title("Cube function")

# Add a smaller set of axes to the figure (inset plot)
axes2 = figure.add_axes([0.6, 0.2, 0.3, 0.3])  
# Smaller subplot within the main plot
axes2.plot(x, z, "r")  # Plot the fourth power function (z = x^4) in red ("r")

# Label the axes and set a title for the inset plot
axes2.set_xlabel("X values")
axes2.set_ylabel("Y values")
axes2.set_title("Square function")  

# Display the figure with both plots
plt.show()
```

```python
# EXAMPLE 2 - Multiple plots on the same axes with a legend

# Create a new figure for plotting
figure = plt.figure(figsize=(6,3))

# Add axes to the figure (occupies the full figure space)
axes = figure.add_axes([0, 0, 1, 1])

# Plot the fourth power function (z = x^4) and label it as "square"
axes.plot(x, z, label="square")

# Plot the cube function (y = x^3) and label it as "cube"
axes.plot(x, y, label="cube")

# Add a legend to distinguish the two lines, placed in the lower right corne
plt.legend(loc=4)
plt.show()# Display the plot
```

```python
# EXAMPLE 3 - Subplots in a grid layout, saving the figure

# Create a 2x1 grid of subplots with a figure size of 6x6 inches
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(6, 3))

# Plot the cube function (y = x^3) on the first subplot (axes[0])
axes[0].plot(x, y)
axes[0].set_title("plot 1")  # Set the title for the first plot

# Plot the fourth power function (z = x^4) on the second subplot (axes[1])
axes[1].plot(x, z)
axes[1].set_title("plot 2")  # Set the title for the second plot

plt.tight_layout() # Automatically adjust the layout to avoid overlapping 

fig.savefig("figure.png") # Save the entire figure as an image file 

plt.show() # Show the figure with the two subplots
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# EXAMPLE 1 - Basic Plot with Two Curves

x = np.arange(-10, 11)
plt.figure(figsize=(6, 3))  # Set figure size

plt.title('My Nice Plot')  # Add title

# Plotting x^2 and -x^2
plt.plot(x, x ** 2)  # Positive parabola
plt.plot(x, -1 * (x ** 2))  # Negative parabola
plt.show()
```

```python
# EXAMPLE 2 - Subplotting with Legends and Gridlines

plt.figure(figsize=(8, 2.5))  # Set figure size
plt.title('My Nice Plot')

plt.subplot(1, 2, 1)  # Create the first subplot (1 row, 2 columns, 1st plot)
plt.plot(x, x ** 2)  # Positive parabola
plt.plot([0, 0, 0], [-10, 0, 100])  # Vertical line at x=0
plt.legend(['X^2', 'Vertical Line'])  # Add legend
plt.xlabel('X')  # X-axis label
plt.ylabel('X Squared')  # Y-axis label

plt.subplot(1, 2, 2)  # Create the second subplot (1 row, 2 columns, 2nd plot)
plt.plot(x, -1 * (x ** 2))  # Negative parabola
plt.plot([-10, 0, 10], [-50, -50, -50])  # Horizontal line at y=-50
plt.legend(['-X^2', 'Horizontal Line'])  # Add legend

plt.xlabel('X')  # X-axis label
plt.ylabel('X Squared')  # Y-axis label
plt.show()
```

```python
# EXAMPLE 3 - Customized Plot with Markers and Labels

fig, axes = plt.subplots(figsize=(6, 3))  # Create figure and axes

axes.plot(x, (x ** 2), color='purple', linewidth=3, marker='o', 
          markersize=8, label='X^2')  # X^2 curve with red color and markers
axes.plot(x, -1 * (x ** 2), 'b--', label='-X^2')  # Dashed blue line for -X^2

axes.set_xlabel('X')  # X-axis label
axes.set_ylabel('X Squared')  # Y-axis label
axes.set_title("My Nice Plot")  # Add title
axes.legend()  # Add legend

plt.show()
```

```python
# EXAMPLE 4 - Plot with Different Linestyles

fig, axes = plt.subplots(figsize=(6,3))  # Set figure size

# Plot lines with different styles
axes.plot(x, x - 0, linestyle='solid')  # Solid line
axes.plot(x, x + 1, linestyle='dashed')  # Dashed line
axes.plot(x, x ** 2, linestyle='dashdot')  # Dashdot line
axes.plot(x, x * 3, linestyle='dotted')  # Dotted line

axes.set_title("My Nice Plot")  # Add title
plt.show()
```

```python
# EXAMPLE 5 - Multiple Subplots in a Grid

plot_objects = plt.subplots(nrows=2, ncols=2, figsize=(6, 3))  
# Create 2x2 subplots grid

fig, ((ax1, ax2), (ax3, ax4)) = plot_objects  # Unpack axes

# Plot different random data in each subplot
ax1.plot(np.random.randn(50), c='purple', linestyle='--')
ax2.plot(np.random.randn(50), c='orange', linestyle=':')
ax3.plot(np.random.randn(50), c='blue', marker='o', linewidth=3.0)
ax4.plot(np.random.randn(50), c='yellow')

plt.show()
```

```python
# EXAMPLE 6 - Grid-based Subplotting

plt.figure(figsize=(8,4))  # Set figure size

# Create subplots using grid layout
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)  # Large subplot
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=1)
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))

plt.show()
```

```python
# EXAMPLE 7 - Scatter Plot with Random Data

N = 50  # Number of points
x = np.random.rand(N)  # Random x coordinates
y = np.random.rand(N)  # Random y coordinates
colors = np.random.rand(N)  # Random colors
area = np.pi * (20 * np.random.rand(N))**2  # Random sizes for points

plt.figure(figsize=(6, 3))  # Set figure size

plt.scatter(x, y, s=area, c=colors, alpha=0.5, 
            cmap='Spectral')  # Create scatter plot
plt.colorbar()  # Add colorbar
plt.show()
```

```python
# EXAMPLE 8 - Multiple Scatter Plots in a Figure
N = 50
x = np.random.rand(N)  # Random x coordinates
y = np.random.rand(N)  # Random y coordinates
colors = np.random.rand(N)  # Random colors
area = np.pi * (20 * np.random.rand(N))**2  
# Random sizes, ensuring same size as x and y

fig = plt.figure(figsize=(14, 4))  # Set figure size

# First scatter plot
ax1 = fig.add_subplot(1, 2, 1) # Scatter plot 1
scatter1 = ax1.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel1')  
plt.colorbar(scatter1)  # Add colorbar to the first plot

# Second scatter plot
ax2 = fig.add_subplot(1, 2, 2) # Scatter plot 2
scatter2 = ax2.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')  
plt.colorbar(scatter2)  # Add colorbar to the second plot

plt.show()
```

```python
# EXAMPLE 9 - Histogram

values = np.random.randn(1000)  # Generate 1000 random values

plt.subplots(figsize=(6,3))  # Set figure size

plt.hist(values, bins=130, alpha=0.8, histtype='bar', 
         color='steelblue', edgecolor='green')  # Create histogram
plt.xlim(xmin=-5, xmax=5)  # Set x-axis limits
plt.show()

# Save the figure as a file
fig.savefig('hist.png')
```

```python
# EXAMPLE 10 - KDE (Kernel Density Estimation)

from scipy import stats

density = stats.kde.gaussian_kde(values)  # Compute KDE for values

plt.subplots(figsize=(6,3))  # Set figure size

values2 = np.linspace(min(values)-10, max(values)+10, 100)  
# Generate x values for plotting KDE

plt.plot(values2, density(values2), color='#FF7F00')  # Plot KDE
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='#FF7F00')  
# Fill area under the curve
plt.xlim(xmin=-5, xmax=5)  # Set x-axis limits
plt.show()
```

```python
# EXAMPLE 11 - Histogram and KDE Combined

plt.subplots(figsize=(6,3))  # Set figure size

# Plot histogram with normalized values
plt.hist(values, bins=100, alpha=0.8, density=1, histtype='bar', 
         color='steelblue', edgecolor='green')

# Overlay KDE on the histogram
plt.plot(values2, density(values2), color='#FF7F00', linewidth=3.0)
plt.xlim(xmin=-5, xmax=5)
plt.show()
```

```python
# EXAMPLE 12 - Stacked Bar Chart

plt.figure(figsize=(6,3))  # Set figure size

Y = np.random.rand(1, 5)[0]  # Random data for first set
Y2 = np.random.rand(1, 5)[0]  # Random data for second set
barWidth = 0.5  # Set width for bars

# Stacked bar chart
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='orange', label='Label Y')
plt.bar(np.arange(len(Y2)), Y2, width=barWidth, color='pink', 
        bottom=Y, label='Label Y2')

plt.legend()  # Add legend
plt.show()
```
