# Text Cleaning

```python
from nltk.stem.porter import PorterStemmer  
# Importing PorterStemmer to perform stemming 
# (reducing words to their root form)
from nltk.corpus import stopwords           
# Importing stopwords (common words like 'the', 'is', etc.)
import re                                   
# Importing regex module for text preprocessing 
```

```python
# Initialize an empty list to store the cleaned and processed comments
corpus = [] 

for k in range(0, 200):  
    # Iterate through the first 200 rows of the dataframe
    ps = PorterStemmer()  
    # Create an instance of PorterStemmer to apply stemming on each comment
    
    # Use regular expressions to keep only alphabetic characters 
    comment = re.sub("[^a-zA-Z]", " ", df["Comments"][k])
    
    # Convert the comment to lowercase and split it into individual words
    comment = comment.lower().split()
    
    total_stopwords = stopwords.words("english")
    total_stopwords.remove("not")
    
    # Stemming: apply stemming to each word in the comment not a stopword
    # Stopwords are common words like 'the', 'and', etc.
    comment = [ps.stem(word) for word in comment if not word 
               in set(total_stopwords)]
    
    # Join the stemmed words back together into a single string
    comment = " ".join(comment)
    
    # Append the processed comment to the 'corpus' list
    corpus.append(comment)
```

```python
print(corpus)
```

# Bag Creating for Words Model

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 550)
```

```python
X = cv.fit_transform(corpus).toarray()
X
```

```python
y = df.iloc[:, -1].values
y
```

# Training the Naive Bayes Model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, 
                                                    random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf = GaussianNB()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), 
                      y_test.reshape(len(y_test), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
