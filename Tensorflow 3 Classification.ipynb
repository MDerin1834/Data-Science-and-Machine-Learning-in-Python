```python
y = df["Type"].values
x = df.drop("Type",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,
                                                    random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

# Tensorflow 

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),
          verbose=1)
```

```python
loss_val = pd.DataFrame(model.history.history)
```

```python
loss_val.plot() # OVERFITTING
```

```python
"EarlyStopping can be used to prevent an increase in loss_val!!!
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
earlyStop = EarlyStopping(monitor="val_loss",mode="min",verbose=1, patience=25)
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),
          verbose=1,callbacks=[earlyStop])
```

```python
Loss_with_Early_Stop = pd.DataFrame(model.history.history)
Loss_with_Early_Stop.plot()
```

# Dropout Usage

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStop])
```

```python
loss_data = pd.DataFrame(model.history.history)
loss_data.plot()
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

```python
predictions = model.predict(x_test)  
# Use predict() instead of predict_classes

# Convert the probabilities to class labels
predicted_classes = np.argmax(predictions, axis=1)

print(predicted_classes)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predicted_classes))
```

```python
print(confusion_matrix(y_test,predicted_classes))
```
