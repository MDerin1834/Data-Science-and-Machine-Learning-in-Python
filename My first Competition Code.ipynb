```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv("used_cars.csv")
df.head()
```

```python
sns.countplot(x = "accident", data= df)
```

```python
df["accident"] = df["accident"].replace({"None reported": 1,
    "At least 1 accident or damage reported": 0})
df.head()
```

```python
df["milage"] = df["milage"].str.replace(" mi.", "").str.replace(",", 
                                                                "").astype(int)
df.head()
```

```python
df["price"] = df["price"].str.replace("$", "").str.replace(",", "").astype(int)
df.head()
```

```python
df.corr()
```

```python
df_clean = df.dropna()
df_clean
```

```python
df_clean = pd.get_dummies(df_clean, columns=["fuel_type","transmission","brand"])
```

```python
df_clean.head()
```

```python
df_clean = df_clean.sort_values(by="price")
df_clean.head()
```

```python
plt.figure(figsize=(7,5))
sns.distplot(df_clean["price"])
```

```python
len(df_clean) * 0.01
```

```python
newCars = df_clean.iloc[33:]
plt.figure(figsize=(7,5))
sns.distplot(df_clean["price"])
```

```python
newCars = newCars.drop(columns=['clean_title','model','int_col','ext_col','engine'])
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression, Lasso, Ridge, ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
train = newCars[:len(newCars)]
X_train = train.drop("price", axis = 1)
y_train = train["price"]
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, 
                                    test_size = 0.3, random_state= 42)
```

# Regression Models

```python
Regressors = [KNeighborsRegressor(), Lasso(), Ridge(), ElasticNet(),
              SVR(), RandomForestRegressor(), GradientBoostingRegressor(),
              LogisticRegression(), DecisionTreeRegressor()]

# Hyperparameters for Grid Search
KNN_param_grid = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}

Lasso_param_grid = {
    'alpha': [0.1, 1.0, 10.0],
    'max_iter': [1000, 5000, 10000],
    'tol': [0.0001, 0.001, 0.01]
}

Ridge_param_grid = {
    'alpha': [0.1, 1.0, 10.0],
    'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg'],
    'max_iter': [1000, 5000, 10000]
}

Elastic_param_grid = {
    'alpha': [0.1, 1.0, 10.0],
    'l1_ratio': [0.1, 0.5, 0.9],
    'max_iter': [1000, 5000, 10000],
    'tol': [0.0001, 0.001, 0.01]}

SVR_param_grid = {
    'C': [0.1, 1.0, 10.0],
    'epsilon': [0.1, 0.2, 0.5],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'degree': [2, 3, 4],  # Only relevant if kernel='poly'
    'gamma': ['scale', 'auto']}

RF_param_grid = {
    'n_estimators': [100, 200, 300],          # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],          # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],          # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4],            # Minimum number of samples required at a leaf node
    'max_features': ['auto', 'sqrt', 'log2'], 
    'bootstrap': [True, False]}

GrBoost_param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5],
    'min_samples_split': [2, 5],
    'subsample': [0.8, 0.9]}

LR_param_grid = {
    'C': [0.01, 0.1, 1.0, 10.0],        # Regularization strength (inverse of regularization)
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],  # Regularization type
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'saga'],  # Solver algorithm
    'max_iter': [100, 200, 500],         # Maximum number of iterations
    'class_weight': [None, 'balanced']}

DT_param_grid = {
    'criterion': ['mse', 'friedman_mse', 'mae', 'poisson'], 
    'max_depth': [None, 10, 20, 30],          # Maximum depth of the tree
    'min_samples_split': [2, 5, 10], # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4],  # Minimum number of samples required at a leaf node
    'max_features': [None, 'auto', 'sqrt', 'log2'],  
    'splitter': ['best', 'random']  # Strategy to choose the split at each node
}
Param_grid_list = [KNN_param_grid,Lasso_param_grid, Ridge_param_grid, 
                   Elastic_param_grid, SVR_param_grid, RF_param_grid,
                  GrBoost_param_grid, LR_param_grid, DT_param_grid]
accuracy_list = []
```

```python
model = KNeighborsRegressor(algorithm= 'ball_tree', 
                            n_neighbors=7, weights= 'distance')
    
# Fit the model to the training data
model.fit(X_train, y_train)
    
# Optionally, make predictions or evaluate the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy_list.append(r2)

print(mse)
print(r2)
```

```python
grid_search = GridSearchCV(estimator = Lasso(), 
                                   param_grid = Lasso_param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best Lasso parameters:",grid_search.best_params_)
```

```python
model2 = Lasso(alpha=10.0, max_iter= 1000, tol= 0.001)
# Fit the model to the training data
model2.fit(X_train, y_train)
    
# Optionally, make predictions or evaluate the model
y_pred = model2.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy_list.append(r2)

print(mse)
print(r2)
```

```python
grid_search = GridSearchCV(estimator = Ridge(), 
                                   param_grid = Ridge_param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best Ridge parameters:",grid_search.best_params_)
```

```python
model3= Ridge(alpha=1.0, max_iter= 1000, solver= 'auto')
# Fit the model to the training data
model3.fit(X_train, y_train)
    
# Optionally, make predictions or evaluate the model
y_pred = model3.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy_list.append(r2)


print(mse)
print(r2)
```

```python
grid_search = GridSearchCV(estimator = ElasticNet(), 
                                   param_grid = Elastic_param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best ElasticNet parameters:",grid_search.best_params_)
```

```python
model4= ElasticNet(alpha= 0.1, l1_ratio= 0.9, max_iter=1000, tol= 0.0001)
# Fit the model to the training data
model4.fit(X_train, y_train)
    
# Optionally, make predictions or evaluate the model
y_pred = model4.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy_list.append(r2)


print(mse)
print(r2)
```

```python
grid_search = GridSearchCV(estimator = GradientBoostingRegressor(), 
                                   param_grid = GrBoost_param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best Gradient Vector parameters:",grid_search.best_params_)
```

```python
model5 = GradientBoostingRegressor(n_estimators= 200, learning_rate=0.3,
    max_depth=2, min_samples_split=5, subsample=0.9)

# Fit the model to the training data
model5.fit(X_train, y_train)
    
# Optionally, make predictions or evaluate the model
y_pred = model5.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy_list.append(r2)

print(mse)
print(r2)
```

```python
model6 = RandomForestRegressor(n_estimators=240,        
    max_depth=20,min_samples_split=5,min_samples_leaf=3)
# Fit the model to the training data
model6.fit(X_train, y_train)
    
# Optionally, make predictions or evaluate the model
y_pred = model6.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy_list.append(r2)

print(mse)
print(r2)
```

```python
model7 = LogisticRegression(C=0.1,max_iter=200)
# Fit the model to the training data
model7.fit(X_train, y_train)
    
# Optionally, make predictions or evaluate the model
y_pred = model7.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy_list.append(r2)

print(mse)
print(r2)
```

```python
model8 = DecisionTreeRegressor(criterion='friedman_mse',max_depth=20,
   min_samples_split=5,min_samples_leaf=2,            
    max_features ='sqrt',  
    splitter='best')
# Fit the model to the training data
model8.fit(X_train, y_train)
    
# Optionally, make predictions or evaluate the model
y_pred = model8.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy_list.append(r2)

print(mse)
print(r2)
```
