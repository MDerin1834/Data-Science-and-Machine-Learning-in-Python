```python
newDf = newDf[newDf.year != 2000]
```

```python
newDf = newDf.drop(["model","transmission","fuelType"],axis=1) # sözel veri atma
```

```python
y = newDf["price"].values
x = newDf.drop("price",axis=1).values  
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,
                        test_size=0.3, random_state=42)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model =  Sequential()

model.add(Dense(15,activation=" "))
model.add(Dense(15,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(15,activation="relu"))

model.add(Dense(1))
model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x = x_train, y = y_train,
          validation_data=(x_test,y_test),
          batch_size=250,epochs=250)
```

```python
KayipVerisi = pd.DataFrame(model.history.history)
KayipVerisi.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
tahmin_dizisi = model.predict(x_test)
tahmin_dizisi
```

```python
mean_absolute_error(y_test, tahmin_dizisi)
```

```python
plt.scatter(y_test,tahmin_dizisi)
plt.plot(y_test,y_test,"r-*")
```
