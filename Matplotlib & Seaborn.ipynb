# SUBPLOTS

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings 
warnings.filterwarnings("ignore")

plt.figure(figsize=(6,3)) # Setting the figure size to 12x6 inches
plt.title("my nice plot") # Adding a title to the overall figure
x = np.linspace(-40, 40) # Creating an array of values from -40 to 40 

plt.subplot(1, 2, 1) # Subplot 1 (1 row, 2 columns, 1st panel selected)
plt.plot(x, x**2) # Plotting x^2 (parabolic curve) for x values
plt.plot([0, 0, 0], [10, 0, 1600]) # Adding a vertical line at x=0 
# (coordinates: [0,0,0] on x-axis, and [10, 0, 1600] on y-axis)
plt.legend(["x^2", "Vertical Line"]) # Adding a legend to explain both lines
plt.xlabel("X") # Labeling the x-axis
plt.ylabel("X square") # Labeling the y-axis

plt.subplot(1, 2, 2) # Subplot 2 (1 row, 2 columns, 2nd panel selected)
plt.plot(x, -1 * x**3) # Plotting -x^3 (inverted cubic curve) for x values
plt.plot([-40, 0, 40], [0, 0, 0]) # Adding a horizontal line at y=0 
# (coordinates: [-40, 0, 40] on x-axis, and [0, 0, 0] on y-axis)
plt.legend(["-x^3", "Horizontal Line"]) # Adding a legend to explain both lines
plt.xlabel("X") # Labeling the x-axis
plt.ylabel("X cube") # Labeling the y-axis

plt.show() # Displaying the entire figure with both subplots
```

```python
# Generating 110 random values between 0 and 1
values = np.random.rand(110)

# Creating a new figure for the histogram plot with size 12x6 inches
plt.subplots(figsize=(8,4))

# Creating a histogram with 100 bin
plt.hist(values, bins=100, alpha=0.8, density=1, color="purple", 
         edgecolor="pink")

# Displaying the histogram plot
plt.show()
```

# 3D PLOTS WITH MATPLOTLIB

```python
ax = plt.axes(projection="3d")
# Create a 3D plot with a scatter projection # Scatter 3D plot

# Single scatter point at coordinates (3, 5, 9)
ax.scatter(3, 5, 9)

# Generating random integer data between 0 and 100 
# for x, y, and z axes (500 points each)
x_data = np.random.randint(0, 100, (500,))
y_data = np.random.randint(0, 100, (500,))
z_data = np.random.randint(0, 100, (500,))

# Scatter plot of the random data with square markers
ax.scatter(x_data, y_data, z_data, marker="s")

# Scatter plot of the random data 
ax.scatter(x_data, y_data, z_data, marker="v", alpha=0.4)

# Show the scatter plots
plt.show()
```

```python
# Line 3D plot
ax = plt.axes(projection="3d")

# Create a range of x and y data from 0 to 100 with a step of 0.1
x_data = np.arange(0, 100, 0.1)
y_data = np.arange(0, 100, 0.1)

# Calculate z as a function of sine(x) * cosine(y) for line plot
z_data = np.sin(x_data) * np.cos(y_data)
# Plot the line in 3D space with x, y, and z data
ax.plot(x_data, y_data, z_data)

# Set the title of the 3D line plot
ax.set_title("Funny Function")
# Label the axes for the line plot
ax.set_xlabel("X values")
ax.set_ylabel("Y values")
ax.set_zlabel("Z values")
# Show the 3D line plot
plt.show()
```

```python
# Surface plot

ax = plt.axes(projection="3d")

# Create a meshgrid for x and y data from 0 to 100 with step size 0.1
x_data = np.arange(0, 100, 0.1)
y_data = np.arange(0, 100, 0.1)
X, Y = np.meshgrid(x_data, y_data)

# Z is calculated as the product of X and Y for the surface plot
Z = X * Y

# Plot the surface using the meshgrid data
ax.plot_surface(X, Y, Z)

plt.show() # Show the surface plot
```

```python
# Second surface plot with a trigonometric function

ax = plt.axes(projection="3d")

# Create a new range for x and y data from -5 to 5 with step size 0.1
x_data = np.arange(-5, 5, 0.1)
y_data = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x_data, y_data)

# Z is calculated using the sine of X multiplied by the cosine of Y
Z = np.sin(X) * np.cos(Y)

# Plot the surface using a "plasma" color map for a visual effect
ax.plot_surface(X, Y, Z, cmap="plasma")

# Show the surface plot with the plasma colormap
plt.show()
```

# PLOTTING WITH SEABORN 

```python
import seaborn as sns

# Create a dictionary with names and grades
data = {"Names": ["John", "Jake", "Jake", "Jake", "Mike", 
        "Mike", "Mike", "Mike", "Mike", "Mike", "Mike", "Mike", "Mike"],
        "Grades": [23, 34, 45, 55, 32, 33, 42, 53, 32, 24, 23, 44, 64]}

# Convert the dictionary into a pandas DataFrame
df = pd.DataFrame(data)
print(df)

# Plot a histogram of the grades
df["Grades"].plot(kind="hist")
plt.show()
```

```python
#%% Scatter Plot using Seaborn
# Load built-in datasets from seaborn
sns.get_dataset_names()

# Load 'tips', 'iris', 'titanic', 'planets' datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")

# Basic scatter plot with 'tip' and 'total_bill' from the tips dataset
sns.scatterplot(x="tip", y="total_bill", data=tips)
```

```python
# Scatter plot with 'day' as hue and 'size' as marker size
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", 
                size="size", palette="YlGnBu")
```

```python
#%% Histogram Plot

# Histogram of 'tip' values with KDE (Kernel Density Estimation)
sns.histplot(tips["tip"], kde=True, bins=50)
```

```python
#%% Bar Plot

# Bar plot showing the average 'tip' based on 'sex'
sns.barplot(x="sex", y="tip", data=tips, palette="YlGnBu")
```

```python
#%% Box Plot

# Box plot of 'total_bill' grouped by 'day'
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
#%% Strip Plot

# Strip plot with 'tip' vs 'day', hue based on 'sex', and a YlGnBu palette
sns.stripplot(x="day", y="tip", data=tips, 
              hue="sex", palette="YlGnBu", dodge=True)
```

```python
# Joint plot (scatter + regression line) of 'tip' and 'total_bill'
sns.jointplot(x="tip", y="total_bill", data=tips, kind="reg")
```

```python
# Joint plot with KDE (Kernel Density Estimation)
sns.jointplot(x="tip", y="total_bill", data=tips, kind="kde", 
              shade=True, cmap="YlGnBu")
```

```python
# Joint plot with hexagonal binning
sns.jointplot(x="tip", y="total_bill", data=tips, kind="hex", cmap="YlGnBu")
```

```python
# Pair plot for numeric data in the titanic dataset, grouped by passenger class
sns.pairplot(titanic.select_dtypes(["number"]), hue="pclass")
```

```python
# Compute correlation matrix of the titanic dataset
print(titanic.corr())
```

```python
# Correlation heatmap with annotations
sns.heatmap(titanic.corr(), annot=True, cmap="coolwarm")
```

```python
#%% Pair Plot for car crashes dataset
# Load car crashes dataset
car_crashes = sns.load_dataset("car_crashes")

# Pair plot for the car crashes dataset
sns.pairplot(car_crashes)
```

```python
# Adjust the context and plot a correlation heatmap for car crashes dataset
plt.figure(figsize=(8, 6))
sns.set_context("paper", font_scale=1.4)
crashes_corr_matrix = car_crashes.corr()
sns.heatmap(crashes_corr_matrix, annot=True, cmap="coolwarm")
```

```python
#%% Heatmap and Cluster Map for Flights Dataset

# Load flights dataset
flights = sns.load_dataset("flights")

# Pivot the flights data to create a matrix of passengers by year and month
flights_pivot = flights.pivot_table(index="month", columns="year", 
                                    values="passengers")
# Heatmap of the flights dataset
sns.heatmap(flights_pivot, cmap="Blues", linecolor="white", linewidth=1)
```

```python
# Cluster map for flights dataset
sns.clustermap(flights_pivot, cmap="Blues", standard_scale=1)
```

```python
#%% Pair Grid for Iris Dataset

# Pair grid plot for iris dataset, grouped by species
iris_grid = sns.PairGrid(iris, hue="species")
iris_grid.map_diag(plt.hist)
iris_grid.map_upper(plt.scatter)
iris_grid.map_lower(sns.kdeplot)
```

```python
# Another Pair Grid with specific x and y variables
iris_grid_2 = sns.PairGrid(iris, hue="species", 
                           x_vars=["sepal_length", "sepal_width"],
                           y_vars=["petal_length", "petal_width"])
iris_grid_2.map(sns.scatterplot)
iris_grid_2.add_legend()
```

```python
#%% Regression Plot
# Set context and plot a regression plot with different markers for each gender
plt.figure(figsize=(7, 4))
sns.set_context("paper", font_scale=1.4)
sns.lmplot(x="total_bill", y="tip", hue="sex", 
           data=tips, markers=["o", "^"], 
           scatter_kws={"s": 100, "linewidth": 0.5, "edgecolor": "w"})
```
